```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine
```

```python
from sklearn.datasets import load_wine

wine_data = load_wine()
X = wine_data.data
y = wine_data.target
```

```python
print(wine_data.DESCR)
```

```python
# Преобразование в DataFrame
df_wine = pd.DataFrame(X, columns=wine_data.feature_names)
df_wine['target'] = y
```

```python
# **Добавление целевой переменной**
df_wine['target'] = y
```

```python
# Проверка наличия пропусков
print(df_wine.isnull().sum())
```

Пропусков нет

```python
# Построение основных графиков для разведочного анализа данных
# Гистограммы распределения признаков
plt.figure(figsize=(15, 10))
for i in range(len(wine_data.feature_names)):
    plt.subplot(4, 4, i+1)
    plt.hist(X[:, i])
    plt.title(wine_data.feature_names[i])
plt.tight_layout()
plt.show()
```

```python
#Визуализации распределения каждого признака в наборе данных "wine" по отношению к целевой переменной
for feature in wine_data.feature_names:
    sns.boxplot(
        x="target",
        y=feature,
        showmeans=True,  # Optional: show means of each class
        data=df_wine
    )
    plt.title(f"{feature} distribution by target class")
    plt.show()
```

```python
#Матрица корреляций
pd_wine_data = pd.DataFrame(wine_data.data, columns=wine_data.feature_names)
pd_wine_data['target'] = wine_data.target

corr = pd_wine_data.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr, annot=True, cmap='YlOrBr')
plt.tight_layout()
plt.show()
```

```python
#Диаграмма рассеяния
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.colorbar()
plt.title("Диаграмма рассеяния")
plt.xlabel(wine_data.feature_names[0])
plt.ylabel(wine_data.feature_names[1])
plt.show()
```

   Мной были построены гистограммы для распределения признаков, коробчатые диаграммы разброса значений, медиану и выбросы для каждой характеристики в зависимости от целевой переменной, матрица корреляций, которая отображает взаимосвязь между всеми парами характеристик. Распределение некоторых характеристик (например, "Alcohol") имеет асимметричную форму. Существуют различия в распределении характеристик между типами вина. Некоторые характеристики (например, "Alcohol" и "Malic acid") имеют сильную положительную корреляцию между собой.
    Диаграмма рассеяния отображает зависимость между двумя произвольными характеристиками (в данном случае, "Alcohol" и "Malic acid"). Цветовая кодировка показывает тип вина для каждого примера.
